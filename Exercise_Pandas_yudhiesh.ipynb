## Question 1

Given the UFO sightings data and the path for csv file, read the csv file from the URL

```python
path ='http://bit.ly/uforeports'
```

use the read_csv to read the csv and assign to a variable called 'ufo'

```python
#read csv
ufo = pd.read_csv(path)
```

```python
#show the top 5 rows of data
ufo.head()
```

```python
#show the last 10 rows of data
ufo.tail(10)
```

```python
#check the data type
ufo.info()
```

```python
#show all rows for the column 'City'
ufo["City"]
```

```python
#determine the shape of the data
ufo.shape
```

```python
#show all data for 'City' that starts with 'E'
ufo[ufo['City'].str[0].isin(['E'])]
```

```python
# count number of reported cases for 'CIRCLE'
# and count the number of reported cases for 'LIGHT'
circle = ufo[ufo["Shape Reported"] == "CIRCLE"].count(axis = 0)
light = ufo[ufo["Shape Reported"] == "LIGHT"].count(axis =0)
print("Number of reported cases for CIRCLE:" ,circle["Shape Reported"])
print("Number of reported cases for LIGHT",light["Shape Reported"])
```

```python
#determine what are the top three colors reported
count = ufo['Colors Reported'].value_counts()
count[:3]
```

```python
#count the number of shape reported by state and city
ufo.groupby(["State", "City"])["Shape Reported"].value_counts()
```

## Question 2

Given the IMDB movies dataset and path, use the read_csv to read the data and assign to a variable 'movies'

```python
path2 = 'http://bit.ly/imdbratings'
```

```python
#read the dataset
movie = pd.read_csv(path2)
```

```python
#show the top 10 rows
movie.head(10)
```

```python
#show summary of the dataset
movie.info()
```

```python
#determine the dimension of the dataset
movie.shape
```

```python
#show the data types for each coloumns
type_ = movie.dtypes
type_
```

```python
#show all the column names
list(movie.columns)
```

```python
# rename the following columns, 'star_rating' as 'stars_rating'
# 'content_rating' as 'content' 

movie.rename(columns={"star_rating": "stars_rating", "content_rating": "content"})
```

```python
#given a list called 'col_names' change all the columns names to 'col_names'
col_names =[]
movie.columns = col_names
```

```python
#provide the statistical summary for 'star_rating' and 'duration'

movie.describe()
```

```python
#show the data type for column 'star_rating'
movie['star_rating'].dtype
```

```python
#show the 5th row data for column 'content_rating'
movie["content_rating"].iloc[5:6]
```

```python
#show all rows for movies duration more than 200 mins
movie[movie["duration"] > 200]
```

```python
#what is the average movie duration
movie["duration"].mean()
```

```python
movie.actors_list.replace(['\[', 'u\'','\'\]'],'', regex=True, inplace=True)
movie.actors_list.replace('\',', ',', regex=True, inplace=True)

actors=pd.DataFrame(movie.actors_list.str.split(',').tolist(), columns = ['actor_1','actor_2','actor_3'])
#actors.head()
movie=pd.concat([movie, actors], axis=1)
movie.head()
```

```python
#count the number of movies where actor 'Charlton Heston' acted in
movie[movie["actors_list"] == "Charles Heston"]
# Does not exits 
```

```python
#what are the top 5 genre movies
count = movie['genre'].value_counts()
count[:5]
```

```python
#which movie has the highest rating
movie.loc[movie['star_rating'].idxmax()]
```

```python
#which movie has the lowest rating
movie.loc[movie['star_rating'].idxmin()]
```

```python
# group by genre and content rating and calculate the mean for duration

movie.groupby(["genre", "content_rating"])["duration"].mean()
```
